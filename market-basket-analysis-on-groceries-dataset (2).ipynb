```python
df['itemDescription'] = df['itemDescription'].str.lower()
df['itemDescription'] = df['itemDescription'].str.strip()
df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

grouped = df.groupby(['Member_number','itemDescription'])['itemDescription'].count().unstack().reset_index().fillna(0).set_index('Member_number')
grouped
```

```python
def encode(x) :
    if x >= 1 :
        return 1
    if x <= 0 :
        return 0
    
    
basket = grouped.applymap(encode)
basket
```

```python
frequent_itemsets = apriori(basket, min_support = 0.05, use_colnames =True)
frequent_itemsets = frequent_itemsets.sort_values(by = 'support', ascending = False)
frequent_itemsets
```
